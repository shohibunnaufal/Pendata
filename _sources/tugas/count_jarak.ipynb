# Menghitung Jarak

```python
import pandas as pd
from sklearn.metrics.pairwise import pairwise_distances

file_url = 'https://drive.google.com/file/d/1Z7juKxTequCoJHukHBQvoNxFz7i9qVNV/view?usp=sharing'

file_id = file_url.split('/')[-2]
dwn_url = "https://drive.google.com/uc?id=" + file_id
df = pd.read_csv(dwn_url, sep=';')
df.head(10)
```

**Penjelasan Tipe Data :**

*   Kolom "Umur" adalah tipe data numerik karena nilainya berupa angka.
*   Kolom "Jenis Kelamin" dan "Status " adalah tipe data kategorical/nominal karena nilainya terdiri dari kategori atau label.
*   Kolom "Jabatan" adalah tipe data kategorical/nominal karena nilainya terdiri dari kategori atau label.
*  Kolom "Memiliki mobil" dan "Memiliki motor" adalah tipe data biner  karena nilainya hanya bisa berupa Ya atau Tidak.

**Code untuk menggabungkan atribut yang berbeda ke dalam matriks ketidaksamaan tunggal**

```python
X = df[[ 'jenis_kelamin','umur', 'status', 'mempunyai mobil','mempunyai motor', 'jabatan']].copy()
X['jenis_kelamin'] = pd.Categorical(X['jenis_kelamin']).codes
X['umur'] = pd.Categorical(X['umur']).codes
X['status'] = pd.Categorical(X['status']).codes
X['mempunyai mobil'] = pd.Categorical(X['mempunyai mobil']).codes
X['mempunyai motor'] = pd.Categorical(X['mempunyai motor']).codes
X['jabatan'] = pd.Categorical(X['jabatan'], categories=['Satpam', 'Dosen'], ordered=True).codes
```

**Code untuk menghitung matriks ketidaksamaan dengan skala interval antara [0,0,1.0]**

```python
dist_matrix = pairwise_distances(X, metric='manhattan') / max(pairwise_distances(X, metric='manhattan').flatten())
```

**Code untuk menampilkan matriks ketidaksamaan**

```python
print(dist_matrix)
```
