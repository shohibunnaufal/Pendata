KNN (K-Nearest Neighbors) adalah salah satu metode machine learning dalam klasifikasi dan regresi. Metode ini memprediksi kelas atau nilai target dari sebuah data baru dengan membandingkannya dengan data-data terdekat di sekitarnya dalam ruang fitur.

Secara singkat, KNN bekerja dengan cara mencari K tetangga terdekat dari data baru, di mana K adalah suatu bilangan bulat yang ditentukan sebelumnya. Kemudian, KNN akan mengambil mayoritas kelas dari K tetangga terdekat tersebut sebagai prediksi kelas untuk data baru (untuk klasifikasi) atau menghitung rata-rata nilai target dari K tetangga terdekat tersebut sebagai prediksi nilai target untuk data baru (untuk regresi).



1. Mengimport library yang diperlukan

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor 
from sklearn.preprocessing import LabelEncoder

from sklearn.impute import KNNImputer
from sklearn.preprocessing import OrdinalEncoder
```

2. Import data kita

```python
df = pd.read_csv('https://gist.githubusercontent.com/shohibunnaufal/16420efa3ecb0d483c7274e9c867dcb9/raw/4fe0bc2e59dbe0a5bb8265ba4a2b22545b58a9b4/heartdisease.csv')

jumlah_baris = df.shape[0]
df.head(jumlah_baris)
```

3. Mengecek nilai yang hilang/kosong pada data

```python
df_copy = df.copy()
df_copy.isnull().sum()
```

4. Ganti nilai yang mengandung tanda tanya (?) menjadi nilai kosong

```python
df_copy = df_copy.replace('?',np.nan)
```

5. Cek lagi berapa nilai yang kosong pada data

```python
df_copy.isnull().sum()
```

6. Setelah itu kita lihat type datanya, jika type datanya bukan numerik/float maka perlu diganti type datanya

```python
df_copy.info()
```

7. Melakukan Normalisasi data mengubah setiap nilai ke dalam rentang 0 hingga 1,dengan metode min-max dan juga metode simple feature scalling(z-score) .namun untuk Kolom target(kolom species ) tak akan di normalisasi

```python
target_column = df_copy['target']
df_copy = df_copy[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']].copy()
df_copy = (df_copy - df_copy.min())/(df_copy.max() - df_copy.min())
df_copy['target'] = target_column.copy()
df_copy.head(jumlah_baris)
```

8.Pada langkah-langkah terakhir dalam inisialisasi objek knn dari kelas KNeighborsRegressor, kita menggunakan algoritma K-Nearest Neighbors untuk melakukan regresi. Untuk memastikan hasil yang konsisten, kita melakukan randomisasi indeks dataset menggunakan np.random.permutation() dengan seed 1.

Kemudian, kita membagi dataset menjadi data training sebesar 75% dan data testing sebesar 25% menggunakan variabel train_index. Data training kemudian disimpan di variabel training_data, sementara data testing disimpan di variabel testing_data.

Setelah itu, kita melakukan training model menggunakan data training dengan memanggil method fit() pada objek knn, dengan argumen input berupa fitur-fitur yang digunakan (dalam hal ini sepal_length, sepal_width, petal_length, dan petal_width) dan label yaitu species.

Setelah model terlatih, kita melakukan prediksi pada data testing dengan memanggil method predict() pada objek knn dengan argumen input berupa fitur-fitur dari data testing yang ingin diprediksi.

Kemudian, untuk mengukur akurasi model, kita menghitung persentase data testing yang berhasil diprediksi dengan menggunakan threshold default sebesar 0.5.

Terakhir, hasil akurasi model dicetak untuk dianalisis lebih lanjut

```python
knn = KNeighborsRegressor()

# before we split, make sure you randomize the dataset
np.random.seed(1)
index_shuffled = np.random.permutation(df_copy.index)
shuffled_car_data = df_copy.loc[index_shuffled]

# We will use 75% of the data for training and the remaining 25% for testing
train_index = int(shuffled_car_data.shape[0] * 0.75)
training_data = shuffled_car_data.iloc[0:train_index]
testing_data = shuffled_car_data.iloc[train_index:]

# fit the model with the training data
knn.fit(training_data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']], training_data['target'])

# predict the target of the testing data
prediction = knn.predict(testing_data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']])

# calculate the accuracy of the model
accuracy = np.mean(np.abs(prediction - testing_data['target']) <= 0.5)

print(f"Akurasi model KNN: {accuracy:.2%}")
```
